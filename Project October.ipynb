```python
x = Loan_data.drop('Personal Loan',axis=1)
```

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.30,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler  
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)  
X_test = scaler.transform(X_test)
```

```python
# Let us fit the model on 30%
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
model_logi = LogisticRegression()
model_logi.fit(X_train, y_train)
yl_predict = model_logi.predict(X_test)
model_logi_score = model_logi.score(X_test, y_test)
print(model_logi_score)
print(metrics.confusion_matrix(y_test, yl_predict))
```

```python
model_logi_score_2 = model_logi.score(X_train, y_train)
```

```python
#Iterating the K Value to find maximum accuracy
```

```python
from matplotlib import pyplot as plt

neighbours = np.arange(1, 40)
train_accuracy_plot = np.empty(len(neighbours))
test_accuracy_plot = np.empty(len(neighbours))
# Loop over different values of k
for i, k in enumerate(neighbours):
    train = []
    test = []
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train)
    train.append(knn.score(X_train,y_train))
    test.append(knn.score(X_test,y_test))
    #Compute accuracy on the training set
    train_accuracy_plot[i] = np.mean(train)
    #Compute accuracy on the testing set
    test_accuracy_plot[i] = np.mean(test)
# Generate plot
plt.title('k-NN: Varying Number of Neighbours')
plt.plot(neighbours, test_accuracy_plot, label = 'Testing Accuracy')
plt.plot(neighbours, train_accuracy_plot, label = 'Training Accuracy')
plt.legend()
plt.xlabel('Number of Neighbours')
plt.ylabel('Accuracy')
plt.show()
```

```python
#I found that at K=3, the model shows maximum accuracy and hence the assumption of K=3 is correct.
```
