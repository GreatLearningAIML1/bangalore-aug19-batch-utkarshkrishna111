```python
Laliga['GoalsFor'] = Laliga['GoalsFor'].astype(int)
Laliga['GoalsAgainst'] = Laliga['GoalsAgainst'].astype(int)
```

```python
def Goal_diff_count():
    GoalsperTeam=Laliga[['Team','GoalsFor','GoalsAgainst']]
    GoalsperTeam=GoalsperTeam.set_index('Team')
    GoalsperTeam['Goalsdiff'] = GoalsperTeam['GoalsFor']-GoalsperTeam['GoalsAgainst']
    GoalsperTeam = GoalsperTeam.sort_values(by= ['Goalsdiff'], ascending = False)
    print(GoalsperTeam)
    print('Maximum Goal Difference and Team')
    print(GoalsperTeam['Goalsdiff'].max())
    print(GoalsperTeam.head(1))
    print('\n')
    print('Minimum Goal Difference and Team')
    print(GoalsperTeam['Goalsdiff'].min())
    print(GoalsperTeam.tail(1))
```

```python
Goal_diff_count()
```
